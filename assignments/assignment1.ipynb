```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)

print(x)
```

```python
xx, yy = np.meshgrid(x,y)
print(yy)
```

```python
plt.pcolormesh(xx)
```

```python
plt.pcolormesh(yy)
```

```python
r = np.sqrt(xx**2 + yy**2)
print(r)
```

```python
theta = np.arctan2(yy, xx)
print(theta)
```

```python
plt.pcolormesh(r)
```

```python
plt.pcolormesh(theta)
```

```python
f = (np.cos(4*r) ** 2) + (np.sin(4*theta)**2)
plt.pcolormesh(f)
```

```python
f_ymean = f.mean(axis=0)
f_xmean = f.mean(axis=1)
```

```python
plt.plot(x, f_ymean)
```

```python
plt.plot(f_xmean, y)
```

```python
import pooch
url = "https://www.ldeo.columbia.edu/~rpa/float_data_4901412.zip"
files = pooch.retrieve(url, processor=pooch.Unzip(), known_hash="2a703c720302c682f1662181d329c9f22f9f10e1539dc2d6082160a469165009")
files
```

```python
date = np.load(files[0])
t = np.load(files[1])
s = np.load(files[2])
p = np.load(files[3])
level = np.load(files[4])
lon = np.load(files[5])
lat = np.load(files[6])
```

```python
pressure_means = []
pressure_std_dev = []
for col in p[:,0:]:
    pressure_means.append(np.nanmean(col))
    pressure_std_dev.append(np.nanstd(col))

print(pressure_means)
print(pressure_std_dev)
```

```python
temp_means = []
temp_std_dev = []
for col in t[:,0:]:
    temp_means.append(np.nanmean(col))
    temp_std_dev.append(np.nanstd(col))

print(temp_means)
print(temp_std_dev)
```
