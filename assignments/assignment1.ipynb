```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)

print(x)
```

```python
xx, yy = np.meshgrid(x,y)
print(yy)
```

```python
plt.pcolormesh(xx)
plt.show()
print(xx.shape)
```

```python
plt.pcolormesh(yy)
plt.show()
```

```python
r = np.sqrt(xx**2 + yy**2)
print(r)
```

```python
theta = np.arctan2(yy, xx)
print(theta)
```

```python
plt.pcolormesh(r)
plt.show()
```

```python
plt.pcolormesh(theta)
plt.show()
```

```python
f = (np.cos(4*r) ** 2) + (np.sin(4*theta)**2)
plt.pcolormesh(f)
plt.show()
```

```python
f_ymean = f.mean(axis=0)
f_xmean = f.mean(axis=1)
```

```python
plt.plot(x, f_ymean)
plt.show()
```

```python
plt.plot(f_xmean, y)
plt.show()
```

# Part 2: Analyze ARGO Data

```python
import pooch
url = "https://www.ldeo.columbia.edu/~rpa/float_data_4901412.zip"
files = pooch.retrieve(url, processor=pooch.Unzip(), known_hash="2a703c720302c682f1662181d329c9f22f9f10e1539dc2d6082160a469165009")
files
```

```python
date = np.load(files[0])
t = np.load(files[1]) # Temperature
s = np.load(files[2]) # Salinity
p = np.load(files[3]) # Pressure
level = np.load(files[4])
lon = np.load(files[5])
lat = np.load(files[6])
```

```python
print(t.shape)
print(s.shape)
print(p.shape)
print('\n')
print(lon.shape)
print(lat.shape)
print(date.shape)
print(level.shape)
```

```python
plt.plot(t[:,0:])
plt.show()
```

```python
plt.plot(s[:,0:])
plt.show()
```

```python

plt.plot(p[:,0:])
plt.title('Pressure')
plt.show()
```

```python
pressure_means = []
pressure_std_dev = []
for col in p[:,0:]:
    pressure_means.append(np.nanmean(col))
    pressure_std_dev.append(np.nanstd(col))

print(pressure_means)
print(pressure_std_dev)
```

```python
temp_means = []
temp_std_dev = []
for col in t[:,0:]:
    temp_means.append(np.nanmean(col))
    temp_std_dev.append(np.nanstd(col))

print(temp_means)
print(temp_std_dev)
```

```python
salinity_means = []
salinity_std_dev = []
for col in s[:,0:]:
    salinity_means.append(np.nanmean(col))
    salinity_std_dev.append(np.nanstd(col))

print(salinity_means)
print(salinity_std_dev)
```

```python
plt.plot(salinity_means, level)
plt.title('Salinity Mean')
plt.show()
```

```python
plt.plot(temp_means,level)
plt.title("Temperature Mean")
plt.show()
```

```python
plt.plot(pressure_means, level)
```

```python
depth_levels = range(1, len(salinity_means) + 1)

# Plotting the line graph with error bars
plt.errorbar(depth_levels, salinity_means, yerr=salinity_std_dev, fmt='o-', label='Salinity Mean')

# Adding labels and title
plt.xlabel('Depth Levels')
plt.ylabel('Salinity Mean')
plt.title('Salinity Mean with Error Bars')
plt.legend()

# Show the plot
plt.show()
```

```python
depth_levels = range(1, len(temp_means) + 1)

# Plotting the line graph with error bars
plt.errorbar(depth_levels, temp_means, yerr=temp_std_dev, fmt='o-', label='Temperature Mean')

# Adding labels and title
plt.xlabel('Depth Levels')
plt.ylabel('Temperature Mean')
plt.title('Temperature Mean with Error Bars')
plt.legend()

# Show the plot
plt.show()
```

```python
depth_levels = range(1, len(pressure_means) + 1)

# Plotting the line graph with error bars
plt.errorbar(depth_levels, pressure_means, yerr=pressure_std_dev, fmt='o-', label='Pressure Mean')

# Adding labels and title
plt.xlabel('Depth Levels')
plt.ylabel('Pressure Mean')
plt.title('Pressure Mean with Error Bars')
plt.legend()

# Show the plot
plt.show()
```
